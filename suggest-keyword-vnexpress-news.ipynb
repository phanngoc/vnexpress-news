```python
df1['Keywords_split'] = df1['keywords'].str.split(',')
df_exploded = df1.explode('Keywords_split')

# Print the exploded DataFrame
print(df_exploded.head(5))
```

```python
# Group by 'Category' and calculate the sum, mean, and count of 'Value' and 'Quantity'
grouped_df = df_exploded.groupby('Keywords_split')['title'].nunique()
# Rename the column without a name
grouped_df.columns = ['A', 'B']
# Print the result
print(grouped_df)
# Sort the values in descending order
sorted_series_desc = grouped_df.sort_values(ascending=False)
print('sorted_series_desc', type(sorted_series_desc))

```

```python
from transformers import AutoModel, AutoTokenizer
model_name = "trituenhantaoio/bert-base-vietnamese-uncased"

phobert = AutoModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
import torch
import numpy as np
from sklearn.cluster import KMeans
from transformers import BertTokenizer, BertModel
from torch.nn.utils.rnn import pad_sequence
from pprint import pprint

# Step 1: Data Preparation
keywords = grouped_df.index.values

tokenized_keywords = []
for keyword in keywords:
    tokens = tokenizer.tokenize(keyword)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    tokenized_keywords.append(token_ids)

input_ids = pad_sequence([torch.tensor(sublist) for sublist in tokenized_keywords], batch_first=True)
# print('padded_sequence', padded_sequence)

print('tokenized_keywords', tokenized_keywords)
# input_ids = torch.tensor(tokenized_keywords)

with torch.no_grad():
    outputs = phobert(input_ids)
    hiddenSize = outputs[0].size()[0]
    # flatten dimension 1 and 2, and hold batch dimension 0
    embeddings = outputs[0].view(hiddenSize, -1).numpy()
    # embeddings = outputs[0][:, 0, :].numpy()  # Extract embeddings from the first layer


k = 150  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=0)
cluster_labels = kmeans.fit_predict(embeddings)
```

```python
import pickle

with open("./models/kmeans.pkl", "wb") as f:
    pickle.dump(kmeans, f)
```

, thì các keyword đó có khả năng cao là giống nhau.

```python
print('cluster_labels', cluster_labels, len(cluster_labels))
# Step 7: Assign Cluster Labels
keyword_clusters = {}
for i, keyword in enumerate(keywords):
    cluster_label = cluster_labels[i]
    if cluster_label not in keyword_clusters:
        keyword_clusters[cluster_label] = []
    keyword_clusters[cluster_label].append(keyword)

print('keyword_clusters', keyword_clusters)

from itertools import zip_longest

# Determine the maximum length of the values
max_length = max(len(v) for v in keyword_clusters.values())
print('max_length', max_length)
# Pad the values to the maximum length with a specific value (e.g., 'N/A')
padded_values = [v + [None] * (max_length - len(v)) for v in keyword_clusters.values()]
print('padded_values', padded_values)

# Create a new dictionary with padded values
padded_data_dict = dict(zip(keyword_clusters.keys(), zip_longest(*padded_values, fillvalue=None)))
print('padded_data_dict', keyword_clusters.keys())

# Convert the dictionary to a DataFrame
keyword_clusters_df = pd.DataFrame.from_dict(padded_data_dict, orient='index')
print('keyword_clusters_df', keyword_clusters_df)
# print('keyword_clusters_df:inf`o', keyword_clusters_df.info())
```

```python
print('keyword_clusters_df', keyword_clusters_df[18].values)
```

```python
def getClusterByKeyword(keyword):
    for i, cluster in enumerate(keyword_clusters_df.values):
        if keyword in cluster:
            return i
    return -1

t = df1['Keywords_split'].values.tolist()
print('t', t[:2])
results = []
for i in t:
    r = []
    for i1 in i:
        r.append(getClusterByKeyword(i1))
    results.append(r)

print('results', results[:2])


from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
encodingLabels = mlb.fit_transform(results)
print('encodingLabels', encodingLabels[:2])

from pickle import dump
# save the scaler
dump(mlb, open('./models/mlb.pkl', 'wb'))
# save list keywords
dump(results, open('./models/keywords.pkl', 'wb'))
# save dict keyword_clusters
dump(keyword_clusters, open('./models/keyword_clusters.pkl', 'wb'))

print('encodingLabels', len(encodingLabels), len(encodingLabels[0]))
print('df1', df1.shape)

df1['Labels'] = list(encodingLabels)
df1.head(5)
df1.to_csv('./data/news_labeled.csv', index=True)
```
